```python
def assert_in_range(s, lo= None, hi = None, name = "value"):
    x = s.dropna()
    if lo is not None:
        assert (x >= lo).all(), f"{name} below {lo}"
    if hi is not None:
        assert (x <= hi).all, f"{name} above {hi}"
```
